```python
# Functions to work with datashader
cvsopts = dict(plot_height = 600, plot_width = 600)

def nodesplot(nodes, name = None, canvas = None, cat = None):

    canvas = ds.Canvas(**cvsopts) if canvas is None else canvas
    aggregator = None if cat is None else ds.count_cat(cat)
    agg = canvas.points(nodes,'x','y',aggregator)
    return tf.spread(tf.shade(agg, cmap=["#FF3333"]), px = 3, name = name)

def edgesplot(edges, name = None, canvas = None):

    canvas = ds.Canvas(**cvsopts) if canvas is None else canvas
    return tf.shade(canvas.line(edges, 'x','y', agg = ds.count()), name = name)

def graphplot(nodes, edges, name = "", canvas = None, cat = None):

    if canvas is None:

        xr = nodes.x.min(), nodes.x.max()
        yr = nodes.y.min(), nodes.y.max()
        canvas = ds.Canvas(x_range = xr, y_range = yr, **cvsopts)

    np = nodesplot(nodes, name + " nodes", canvas, cat)
    ep = edgesplot(edges, name + " edges", canvas)
    return tf.stack(ep, np, how = "over", name = name)
```

```python
# Read the file using pandas
df = pd.read_csv('/content/drive/MyDrive/netflix_titles.csv').drop('show_id', axis = 1)
```

```python
# Split the cells to extract data
df['directors'] = df['director'].apply(lambda l: [] if pd.isna(l) else [i.strip() for i in l.split(",")])
df['categories'] = df['listed_in'].apply(lambda l: [] if pd.isna(l) else [i.strip() for i in l.split(",")])
df['actors'] = df['cast'].apply(lambda l: [] if pd.isna(l) else [i.strip() for i in l.split(",")])
df['countries'] = df['country'].apply(lambda l: [] if pd.isna(l) else [i.strip() for i in l.split(",")])
```

```python
df.head()
```

```python
# Init the graph
graph = nx.Graph(label = 'movie')
```

```python
# Iterate the df to build netwrosk
start_time = time.time()

for i, rowi in df[df['type'] == 'Movie'].iterrows():

    if (i % 1000 == 0):

        print(" iter {} -- {} seconds --".format(i,time.time() - start_time))

    graph.add_node(rowi['title'], label = "MOVIE", mtype = rowi['type'], rating = rowi['rating'])

    # Iterate components to load each
    for element in rowi['actors']:

        graph.add_node(element,label = "PERSON")
        graph.add_edge(rowi['title'], element, label = "ACTED_IN")

    for element in rowi['categories']:

        graph.add_node(element, label = "CAT")
        graph.add_edge(rowi['title'], element, label = "CAT_IN")

    for element in rowi['directors']:

        graph.add_node(element, label = "PERSON")
        graph.add_edge(rowi['title'], element, label = "DIRECTED")

    for element in rowi['countries']:

        graph.add_node(element,label = "COU")
        graph.add_edge(rowi['title'], element, label = "COU_IN")

print(" finish -- {} seconds --".format(time.time() - start_time))
```

```python
# Reformat the diagram for informatibr purpose
print("Initial Node Count: ", graph.number_of_nodes())
print("Initial Edge Count: ", graph.number_of_edges())

min_degree = 10

degree_dict = dict(graph.degree())
remove_nodes = [n for n, deg in degree_dict.items() if deg < min_degree]

graph.remove_nodes_from(remove_nodes)

print("Remaining nodes:", graph.number_of_nodes())
print("Remaining edges:", graph.number_of_edges())
```

```python
# Color nodes by type
nodes2 = pd.DataFrame([{"name": n, "node_category": graph.nodes[n]["label"]} for n in graph.nodes()])
nodes2["node_category"] = nodes2["node_category"].astype("category")
```

```python
colors = [i for i in cc.glasbey[0:7]]

category_colors = {"MOVIE": colors[0],
                   "ACTOR": colors[1],
                   "DIRECTOR": colors[2],
                   "COU": colors[3],
                   "GENRE": colors[4],
                   'CAT': colors[5],
                   'PERSON': colors[6]}

nodes2["color"] = nodes2["node_category"].map(category_colors)

# Add sizing
nodes2["degree"] = nodes2["name"].map(dict(graph.degree()))
nodes2["size"] = np.log1p(nodes2["degree"]) * 5
```

```python
# Sample edges to remove clutter
edges1 = edges.sample(frac = 0.35, random_state = 100)
```

```python
%time forcedirected = forceatlas2_layout(nodes2, edges1)
```

```python
# Redfine the plotting logic
fd = forcedirected

def nodesplot(nodes, name = None, canvas = None, cat = None):

    canvas = ds.Canvas(**cvsopts) if canvas is None else canvas
    aggregator = None if cat is None else ds.count_cat(cat)
    agg = canvas.points(nodes, 'x', 'y', aggregator)

    return tf.spread(tf.shade(agg, color_key = category_colors), px = 3, name = name)

plot = graphplot(forcedirected, connect_edges(fd, edges1), "Force Directed Hollywood Layout", cat = "node_category")
plot1 = graphplot(fd, hammer_bundle(fd, edges1), "Bundled Force-Directed Layout")
```
