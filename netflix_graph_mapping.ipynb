```python
# Functions to work with datashader
cvsopts = dict(plot_height = 1000, plot_width = 1000)

def nodesplot(nodes, name = None, canvas = None, cat = None):

    canvas = ds.Canvas(**cvsopts) if canvas is None else canvas
    aggregator = None if cat is None else ds.count_cat(cat)
    agg = canvas.points(nodes,'x','y',aggregator)
    return tf.spread(tf.shade(agg, cmap=["#FF3333"]), px = 3, name = name)

def edgesplot(edges, name = None, canvas = None):

    canvas = ds.Canvas(**cvsopts) if canvas is None else canvas
    return tf.shade(canvas.line(edges, 'x','y', agg = ds.count()), name = name)

def graphplot(nodes, edges, name = "", canvas = None, cat = None):

    if canvas is None:

        xr = nodes.x.min(), nodes.x.max()
        yr = nodes.y.min(), nodes.y.max()
        canvas = ds.Canvas(x_range = xr, y_range = yr, **cvsopts)

    np = nodesplot(nodes, name + " nodes", canvas, cat)
    ep = edgesplot(edges, name + " edges", canvas)
    return tf.stack(ep, np, how = "over", name = name)
```

```python
# Function to calculate positions
def radial_layer_layout(nodes):

    layers = {"genre": 0.2,
              "movie": 1.0,
              'director': 2.0,
              "actor": 3.0}

    coords = []

    for node_type, group in nodes.groupby("node_type"):

        r = layers[node_type]

        n = len(group)

        angles = np.linspace(0, 2*np.pi, n, endpoint=False)

        x = r * np.cos(angles)
        y = r * np.sin(angles)

        temp = group.copy()
        temp["x"] = x
        temp["y"] = y

        coords.append(temp)

    return pd.concat(coords)
```

```python
# Apply function to get coords
nodes1 = radial_layer_layout(nodes)
```

```python
# A new function to handle the changed schema
def connect_edges(nodes, edges):

    node_pos = nodes.set_index("id")[["x","y"]]

    edges = edges.copy()

    edges["x"] = edges["source"].map(node_pos["x"])
    edges["y"] = edges["source"].map(node_pos["y"])

    edges["x_end"] = edges["target"].map(node_pos["x"])
    edges["y_end"] = edges["target"].map(node_pos["y"])

    return edges
```

```python
# Init the graph
graph = nx.Graph(label = 'movie')
```

```python
# Iterate the df to build netwrosk
start_time = time.time()

for i, rowi in df[df['type'] == 'Movie'].iterrows():

    if (i % 1000 == 0):

        print(" iter {} -- {} seconds --".format(i,time.time() - start_time))

    graph.add_node(rowi['title'], label = "MOVIE", mtype = rowi['type'], rating = rowi['rating'])

    # Iterate components to load each
    for element in rowi['actors']:

        graph.add_node(element,label = "PERSON")
        graph.add_edge(rowi['title'], element, label = "ACTED_IN")

    for element in rowi['categories']:

        graph.add_node(element, label = "CAT")
        graph.add_edge(rowi['title'], element, label = "CAT_IN")

    for element in rowi['directors']:

        graph.add_node(element, label = "PERSON")
        graph.add_edge(rowi['title'], element, label = "DIRECTED")

    for element in rowi['countries']:

        graph.add_node(element,label = "COU")
        graph.add_edge(rowi['title'], element, label = "COU_IN")

print(" finish -- {} seconds --".format(time.time() - start_time))
```
