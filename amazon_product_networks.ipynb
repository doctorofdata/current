```python
# Functions to work with datashader
cvsopts = dict(plot_height=400, plot_width=400)

def nodesplot(nodes, name = None, canvas = None, cat = None):

    canvas = ds.Canvas(**cvsopts) if canvas is None else canvas
    aggregator = None if cat is None else ds.count_cat(cat)
    agg = canvas.points(nodes,'x','y',aggregator)
    return tf.spread(tf.shade(agg, cmap=["#FF3333"]), px=3, name=name)

def edgesplot(edges, name = None, canvas = None):

    canvas = ds.Canvas(**cvsopts) if canvas is None else canvas
    return tf.shade(canvas.line(edges, 'x','y', agg = ds.count()), name = name)

def graphplot(nodes, edges, name = "", canvas = None, cat = None):

    if canvas is None:

        xr = nodes.x.min(), nodes.x.max()
        yr = nodes.y.min(), nodes.y.max()
        canvas = ds.Canvas(x_range = xr, y_range = yr, **cvsopts)

    np = nodesplot(nodes, name + " nodes", canvas, cat)
    ep = edgesplot(edges, name + " edges", canvas)
    return tf.stack(ep, np, how = "over", name = name)
```

```python
# Functions to produce edge bundling
def subdivide_edge(p0, p1, segments):

    return np.linspace(p0, p1, segments)

def edge_compatibility(e1, e2):

    p0, p1 = e1
    q0, q1 = e2

    # angle compatibility
    d1 = p1 - p0
    d2 = q1 - q0

    angle = abs(np.dot(d1, d2) / (np.linalg.norm(d1) * np.linalg.norm(d2) + 1e-9))

    # length compatibility
    l1 = np.linalg.norm(d1)
    l2 = np.linalg.norm(d2)

    length = 2 / (l1/l2 + l2/l1)

    # distance compatibility
    mid1 = (p0 + p1) / 2
    mid2 = (q0 + q1) / 2

    dist = np.linalg.norm(mid1 - mid2)

    distance = 1 / (1 + dist)

    return angle * length * distance

def force_directed_edge_bundling(edges, pos, K = 0.1, iterations = 50, segments = 6, compatibility_threshold = 0.6):

    # convert edges to coordinates
    edge_coords = [(np.array(pos[u]), np.array(pos[v])) for u, v in edges]

    # subdivide edges
    bundled = [subdivide_edge(p0, p1, segments) for p0, p1 in edge_coords]

    # compute compatibility matrix
    compat = np.zeros((len(edges), len(edges)))

    for i in range(len(edges)):

        for j in range(i+1, len(edges)):

            c = edge_compatibility(edge_coords[i], edge_coords[j])

            if c > compatibility_threshold:

                compat[i,j] = compat[j,i] = c

    # iterative relaxation
    for _ in range(iterations):

        new_bundled = []

        for i, edge in enumerate(bundled):

            forces = np.zeros_like(edge)

            for j, other in enumerate(bundled):

                if compat[i,j] > 0:

                    diff = other - edge
                    forces += diff * compat[i,j]

            edge = edge + K * forces
            new_bundled.append(edge)

        bundled = new_bundled

    return bundled
```

```python
# Load the data
real_edges = pd.read_csv('com-amazon.ungraph.txt', sep = "\t", comment = "#", header = None, names = ["source_original","target_original"])
```

```python
real_edges.head()
```

```python
# Read the metadata
products = []

current = {}

# From file
with open("amazon-meta.txt", encoding = "latin1") as f:

    for line in f:

        line = line.strip()

        # start new product
        if line.startswith("Id:"):

            if current:

                products.append(current)

            current = {"id": int(line.split()[1])}

        elif line.startswith("ASIN:"):

            current["asin"] = line.split()[1]

        elif line.startswith("title:"):

            current["title"] = line.replace("title:", "").strip()

        elif line.startswith("group:"):

            current["group"] = line.split()[1]

        elif line.startswith("salesrank:"):

            val = line.split()[1]
            current["salesrank"] = int(val) if val.isdigit() else None

        elif line.startswith("similar:"):

            parts = line.split()

            if len(parts) > 2:

                current["similar"] = parts[2:]

        elif line == "":

            if current:

                products.append(current)
                current = {}

# convert to dataframe
products = pd.DataFrame(products)
```
